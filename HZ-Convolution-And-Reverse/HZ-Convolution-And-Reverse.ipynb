```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa as lr
```

## Define my convolution function

```python
def HZ_convolv(x):
    
    #normalize the input signal x
    max_x = 0 #set the value to find out the amplitude of x
    
    for i in range (len(x)):
        if max_x < np.abs(x[i]):
            max_x = np.abs(x[i])
    
    max_amplitude = 1 #set the normalized max amplitude
    
    for i in range (len(x)):
        x[i] = x[i]*float(max_amplitude)/float(max_x)
    
    
    #reverse the input signal; let the reverse be the impulse response
    h = np.zeros(len(x))
    
    for i in range (len(x)):
        h[i] = x[len(x)-1 - i] #reverse
        
    
    #convolve the input signal with its reverse 
    y = np.convolve(x,h)
    
        
    return y
```

## Test my convolution function
    

```python
#first test my function with a 1kHz_Sinewave
```

```python
sine,sr_sine = lr.load('Test Sine Wave.wav')
plt.plot(sine)
#plt.xlim([0,200])
plt.xlim(0,len(sine))
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("time domain of sinewave")
```

```python
Audio(sine,rate=sr_sine)
```

```python
plt.plot(np.abs(np.fft.fft(sine)))
#plt.xlim([0,2000])
#plt.xlim([15000,25000])
plt.title("frequency domain of sinewave")
plt.xlabel("frequency equivalent")
plt.ylabel("amplitude")
```

```python
plt.plot(HZ_convolv(sine))
#plt.xlim([0,len(HZ_convolv(sine))])
#plt.xlim([2000,3000])
plt.title("time domain of sinewave after convolution")
plt.xlabel("time")
plt.ylabel("amplitude")
```

```python
Audio(HZ_convolv(sine),rate=sr_sine)
```

```python
plt.plot(np.abs(np.fft.fft(HZ_convolv(sine))))
#plt.xlim([0,4000])
#plt.xlim([30000,50000])
plt.title("frequency domain of sinewave after convolution")
plt.xlabel("frequency equivalent")
plt.ylabel("amplitude")
```
